```python
import json
import csv
import pandas as pd
import nltk
from nltk.corpus import stopwords
 
```

```python
data = pd.read_csv("jyapu.csv")
```

```python
list(data)
```

```python
clean_data = data.filter(items = ["hashtags","favorite_count","retweet_count","text","user_location"])
```

```python
clean_data.dropna()
```

```python
clean_data.info()
```

```python
clean_data['favorite_count'].plot(kind = 'hist')
```

```python
#stop word removal
test_data = clean_data.head()
stop = set(stopwords.words('english'))
test_data['clean_text'] = test_data['text'].apply(lambda x:''.join([word for word in x.split() if word not in (stop)]))
print(test_data['clean_text'])
```
